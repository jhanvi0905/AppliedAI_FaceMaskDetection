```python
import os

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score,classification_report
import seaborn as s
from dataProcess import load_data, im_convert, validation_loader
import torch
import numpy as np
import matplotlib.pyplot as plt
from train_model import CNN
```

```python
def generate_sample_preds(test_load, model, path):
    dataiter = iter(test_load)
    images, labels = next(dataiter)
    output = model(images)
    _, preds = torch.max(output, 1)
    fig = plt.figure(figsize=(25, 4))
    classes = ['ClothMask', 'FFP2 Mask', 'FFP2 Mask With Valve', 'Surgical Mask', 'Without Mask']
    for idx in np.arange(5):
        ax = fig.add_subplot(10, 10, idx + 1, xticks=[], yticks=[])
        print(classes[preds[idx]])
        ax.set_title(classes[preds[idx]])
        plt.imsave(path + classes[preds[idx]] + str(idx) + '.png',
                   im_convert(images[idx]))
    Accuracy = accuracy_score(preds, labels)
    print("Accuracy of Current Batch: ", Accuracy)
    confusion = confusion_matrix(preds, labels)
    report = classification_report(labels, preds, labels=[0,1,2,3,4], target_names=classes)
    print(report)
    return confusion
```

```python
test_loader1 = load_data("output_fol1/test/")
test_loader2 = load_data("output_fol2/test/")
test_loader_female = validation_loader(test_loader1, 128)
test_loader_male = validation_loader(test_loader2, 128)
test_loader_total = validation_loader(torch.utils.data.ConcatDataset([test_loader1, test_loader2]), 256)
model = torch.load('/Users/jhanviarora/Desktop/Project/output_models/final.h5', map_location=torch.device('cpu'))
confusion_female = generate_sample_preds(test_loader_female, model, '/Users/jhanviarora/Desktop/Project/Output_Predictions/')
confusion_male = generate_sample_preds(test_loader_male, model, '/Users/jhanviarora/Desktop/Project/Output_Predictions/')
confusion_total = generate_sample_preds(test_loader_total, model, '/Users/jhanviarora/Desktop/Project/Output_Predictions/')
```

```python
ax = s.heatmap(confusion_female, annot=True,cmap = "Blues")

ax.set_title('Old Confusion Matrix ')
ax.set_xlabel('Predicted Values')
ax.set_ylabel('Actual Values')
plt.show()
```
